```python
#Import Phonological Transcript
Phon_File = 'br-phono.txt'
```

```python
handle = open (Phon_File, 'r')
phon_data = handle.read()
print(phon_data)
```

```python
phon_words = phon_data.split()
print(phon_words)
```

```python
phon_count = Counter(phon_words)
```

```python
def pdist(counter):
    N = sum(counter.values())
    return lambda x: counter[x]/N
```

```python
P = pdist(phon_count)
```

```python
phon_count_data =[]
for x in phon_count:
    phon_count_data.append([x,P(x)])
import pandas as pd
df = pd.DataFrame(phon_count_data, columns=["Phonic Word", "Probability Distribution"])
df
```

```python
def splits (text, start=0, L=50):
    return [(text[:i], text[i:])
           for i in range(start, min(len(text), L)+1)]
```

```python
def segment(text):
    if not text:
        return[]
    else:
        potential= ([first] + segment (rest) for (first, rest) in splits(text,1))
        return max(potential, key=Pwords)
```

```python
def product(nums):
    result = 1
    for x in nums:
        result *= x
    return result
```

```python
def Pwords(words):
    return product(P(w) for w in words)
```

```python
counter = 0;
compared_data =[];
for x in algo_phon_line:
    phon_compare = 0;    
    counter_2 = 0;
    for y in x: 
        if phon_line_check[counter][counter_2]==y:
            phon_compare += 1;
        counter_2 += 1;
    compared_data.append([phon_line_check[counter],x,(phon_compare/len(x))])
    counter += 1;
```
